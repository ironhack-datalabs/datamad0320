```python
import json
import os
from dotenv import load_dotenv
import requests
import re
```

```python
def getFromGithub(path,queryParams=dict(),apiKey=""):
    
    # Construct the resource url
    url = f"https://api.github.com{path}"
    
    # If apiKey is defined, pass a header
    headers = {"Authorization":f"token {apiKey}"} if apiKey else {}

    # Perform the request
    res = requests.get(url, params=queryParams, headers=headers)
    print(res.status_code, res.url)
    
    # Extract json from body response
    return res.json()
```

```python
load_dotenv()
# Load the apikey
apiKey = os.getenv("GITHUB_APIKEY")
print("APIKEY FOUND!!") if apiKey else print("ERROR: NO APIKEY FOUND")
```

```python
repofiles = getFromGithub("/repos/ironhack-datalabs/scavenger", apiKey=apiKey)
```

```python
def searchPath(path):
    paths =[(getFromGithub(f"/repos/ironhack-datalabs/scavenger/contents/{p}", apiKey=apiKey))for p in path]
    path =[[e["path"] for e in paths[i]]for i in range(1,len(paths))]
    return path
```

```python
search=searchPath(path)
```

```python
filespath = [e for x in search for e in x if re.search(r"[.s]?[t]",e)]
print(filespath)
```

```python
na= [e.split("/") for e in filespath]
name=sorted([e for x in na for e in x if re.search(r"[.s]?[t]",e)])
```
