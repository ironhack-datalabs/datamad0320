```python
# Challenge 1

df=pd.read_sql_query("""

SELECT a.au_id AS AUTHORID, a.au_fname AS LASTNAME, a.au_lname AS FIRSTNAME, t.title AS TITLE, p.pub_name AS PUBLISHER
     FROM authors AS a
         INNER JOIN titleauthor as ta
         ON a.au_id=ta.au_id
         INNER JOIN titles as t
         ON ta.title_id=t.title_id
         INNER JOIN publishers as p
         ON t.pub_id=p.pub_id

""", engine)
df
```

```python
# Challenge 2

df=pd.read_sql_query("""

SELECT a.au_id AS AUTHORID, a.au_fname AS LASTNAME, a.au_lname AS FIRSTNAME, p.pub_name AS PUBLISHER,count(*) AS TITLECOUNT 
     FROM authors AS a
         INNER JOIN titleauthor as ta
         ON a.au_id=ta.au_id
         INNER JOIN titles as t
         ON ta.title_id=t.title_id
         INNER JOIN publishers as p
         ON t.pub_id=p.pub_id
	GROUP BY a.au_id,a.au_fname,a.au_lname,p.pub_name
    ORDER BY TITLECOUNT DESC
	

""", engine)
df
```

```python
# Challenge 3

df=pd.read_sql_query("""

SELECT a.au_id AS AUTHORID, a.au_fname AS LASTNAME, a.au_lname AS FIRSTNAME,sum(s.qty) AS TOTAL
	FROM authors AS a
         INNER JOIN titleauthor as ta
         ON a.au_id=ta.au_id
         INNER JOIN titles as t
         ON ta.title_id=t.title_id
         INNER JOIN sales as s
         ON t.title_id=s.title_id
	GROUP BY a.au_id,a.au_fname,a.au_lname
    ORDER BY TOTAL DESC
   LIMIT 3

	

""", engine)
df
```
