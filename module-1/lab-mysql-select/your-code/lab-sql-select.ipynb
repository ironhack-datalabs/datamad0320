# Challenge 1

```python

df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname AS "FIRST NAME", t.title AS TITLE, pb.pub_name AS PUBLISHER

	FROM titles AS t 
    INNER JOIN titleauthor AS tta
    ON  t.title_id = tta.title_id
    INNER JOIN authors AS au
    ON tta.au_id = au.au_id
    INNER JOIN publishers AS pb
    ON t.pub_id = pb.pub_id
    ORDER BY au.au_id

""", engine)
```

# Challenge 2

```python
df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname AS "FIRST NAME", pb.pub_name AS PUBLISHER, count(t.title) AS "TITLE COUNT"

	FROM titles AS t 
    INNER JOIN titleauthor AS tta
    ON  t.title_id = tta.title_id
    INNER JOIN authors AS au
    ON tta.au_id = au.au_id
    INNER JOIN publishers AS pb
    ON t.pub_id = pb.pub_id
    GROUP BY au.au_id,au.au_lname,au.au_fname,pb.pub_name
    ORDER BY count(t.title)  DESC


""", engine)
```

```python
df
```

# Challenge 3

```python
df = pd.read_sql_query("""
SELECT au.au_id AS "AUTHOR ID", au.au_lname AS "LAST NAME", au.au_fname AS "FIRST NAME", pb.pub_name AS PUBLISHER, sum(s.qty) AS "TOTAL"

	FROM titles AS t 
    INNER JOIN titleauthor AS tta
    ON  t.title_id = tta.title_id
    INNER JOIN authors AS au
    ON tta.au_id = au.au_id
    INNER JOIN publishers AS pb
    ON t.pub_id = pb.pub_id
    INNER JOIN sales as s
    ON t.title_id=s.title_id
    GROUP BY au.au_id,au.au_lname,au.au_fname,pb.pub_name
    ORDER BY sum(s.qty) DESC LIMIT 3

""", engine)
```
