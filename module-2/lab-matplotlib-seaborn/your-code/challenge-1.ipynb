```python
# import libraries here
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

#### Define data.

```python
x = np.arange(0,100)
y = x*2
z = x**2
print(x,"\n", y)
```

#### Plot (x,y) and (x,z) on the axes.

#### There are 2 ways of doing this. Do in both ways.

*Hint: Check out the `nrows`, `ncols`, and `index` arguments of [subplots](https://matplotlib.org/api/_as_gen/matplotlib.pyplot.subplot.html)*

#### Also, play around with the linewidth and style. Use the ones you're most happy with.

```python
# your code here-1st way (call `subplots` twice using the `index` parameter)
plt.figure()

#plt.subplot(1 = 1 linea, 2 = 2 columnas, 1 = lugar columna 1 de 2)
plt.subplot(1,2,1)
plt.plot(x,y, lw=5, color="r", ls="dashdot")
plt.title("Subplot 1")

plt.subplot(1,2,2)
plt.plot(x,z)
plt.title("Subplot 2")
```

```python
# your code here-2nd way (call `subplots` only once not using the `index` parameter)
fig, axes = plt.subplots(1,2)

axes[0].plot(x,y)
axes[0].set_title("Subplot 1") 

axes[1].plot(x,z, lw=10, color="orange", ls="--")
axes[1].set_title("Subplot 2")
```

#### Plot both `y=x^2` and `y=exp(x)`in the same plot using normal and logarithmic scale.

*Hint: Use `set_xscale` and `set_yscale`*

```python
#https://dfrieds.com/data-visualizations/when-use-log-scale.html
# your code here
y1 = x**2
y2 = np.exp(x)
print(y1,"\n",y2)
```

```python
#incluso dentro de esta "manera 2" de hacer los subplots, tenemos una submanera para hacer los subplots
fig, ax = plt.subplots(1,2,figsize=(15,5))

#ambas lineas (x,y) en el mismo plot. El gran problema de esta opt es la customizacion de ls, lw.. Aqui aplica ambas
ax[0].plot(x,y1,x,y2)
#si usas la linea inferior veras que el efecto log en x es practicamente nulo, ya que es en y donde existen 
#los grandes saltos
#ax[0].set_xscale("log")
ax[0].set_title("Normal scale")

#primero un plot y luego otro
ax[1].plot(x,y1)
ax[1].plot(x,y2)
#la escala log es muy util para comparar valores con grandes diferencias
#como podemos ver en el subplot, no vemos casi el crecimiento de las lineas, mientras que el subplot 2 es claro
#hint: siempre que haya grandes rangos de y1/x1 --> yn/xn usa log scale
ax[1].set_yscale("log")
ax[1].set_title("Log scale")
```

Scatter Plot

Please provide a scatter plot between "Combined MPG" as X variable and 
"Highway MPG" as Y variable

```python
# your code here
df = pd.read_csv("vehicles.csv")
print(df.shape)
df.head()
```

```python
df.dtypes
```

```python
#https://matplotlib.org/3.2.1/api/_as_gen/matplotlib.pyplot.scatter.html
x = df["Combined MPG"]
y= df["Highway MPG"]

plt.figure(figsize=(15,5))

plt.scatter(x,y, lw=.5, alpha=1, marker="4")
plt.xlabel("Combined MPG")
plt.ylabel("Highway MPG")
plt.title("Scatter")
```

Como vemos, se pueden apreciar varias diferencias entre matplotlib y seaborn a la hora de representar df:

1)La diferencia de codigo 1 linea vs x.

2)La visibilad de grafico de seaborn es mas clara (separa mejor los puntos, para pyplot he tenido que modificar sus parametros.

3)La forma, mientras que en matplotlib puedes elergir darle forma, seaborn es cuadrado.

4)Seaborn me da ambas columbas como un histogram como aadido.

5)Te identifica los labels de base.

6)Kind; aunque al estar usando jointplot, son todo scatters, podemos hacer una regresion facilmente
vamos a incluirla as verla.

```python
#with use sns.set function to have an according style with statistic o calc data
#veremos que de aqui en adelante incluso los graficos de matplotlib salen con este estilo
#https://seaborn.pydata.org/generated/seaborn.jointplot.html

sns.set()
sns.jointplot("Combined MPG","Highway MPG", data=df, kind="reg", height=7, dropna=True)
```

```python
#tercera opcion usando tb sns, y usando scatter en lugar de jointplot
#como podemos ver los puntos son mas claros
sns.scatterplot("Combined MPG","Highway MPG", data=df)
```

Box Whisker Plot

Please provide a box plot of the variable "CO2 Emission Grams/mile"

```python
# your code here
#https://matplotlib.org/3.1.1/api/_as_gen/matplotlib.pyplot.boxplot.html
plt.figure(figsize=(15,5))

plt.boxplot(df["CO2 Emission Grams/Mile"], vert=False, patch_artist=True, showmeans=True)
plt.xlabel("CO2 Emission Grams/Mile")
plt.title("Box Whisker Plot")

#como podemos ver hemos cambiado a horizontal el boxplot (al ser una variable es mas facil de leer)
#lo estilizamos con patch y vemos la media con showmeans (triangulo sober la mediana)
#tb podemos ver que se ha aplicado el estilo sns.set()
```

```python
#https://seaborn.pydata.org/generated/seaborn.boxplot.html
sns.boxplot("CO2 Emission Grams/Mile", data=df, orient="h")
```

Histogram

Please provide a histogram of the Fuel Barrels/Year

```python
# your code here
plt.figure(figsize=(15,5))

plt.hist(df["Fuel Barrels/Year"], histtype="bar",bins=50)
plt.xlabel("Fuel Barrels/Year")
plt.title("Histogram")
```

Bar Chart

Please provide a bar chart of the Fuel Type on the X axis and "City MPG" on the Y axis

```python
# your code here
#hemos utilizado una escala log, ya que si no la mayoria de los fuel type no se verian
group = df["City MPG"].groupby(df["Fuel Type"]).sum()

plt.figure(figsize=(15,5))
group.plot.bar()
plt.xlabel("Fuel Type")
plt.ylabel("City MPG")
plt.title("Bar Chart")
plt.yscale("log")
print(group,type(group))
```
